```python
import zipfile
zip_file_path = '/content/API_SP.POP.TOTL_DS2_en_csv_v2_38144.zip'

# Open ZIP and inspect the file names
with zipfile.ZipFile(zip_file_path, 'r') as z:
    # List files inside ZIP
    print("Files inside ZIP:", z.namelist())

    # Extract and read the main CSV file
    csv_file_name = 'API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv'  # Make sure this matches
    with z.open(csv_file_name) as f:
        # Preview first 20 lines to locate headers
        for i in range(20):
            print(f.readline().decode('utf-8').strip())
```

The dataset was imported successfully for visualization. The datset is CSV file which contains the population of different  countries.

```python
# Load the CSV file into a pandas DataFrame, skipping the initial metadata rows
with zipfile.ZipFile(zip_file_path, 'r') as z:
    with z.open(csv_file_name) as f:
        df = pd.read_csv(f, skiprows=4)

display(df.head())
```

Now we go with the cleaning and selecting the neccesary columns required for the task and repersent the continuous and categorical variables in the bar and histogram charts.

```python
#data cleaning AND selecting necessary columns
```

```python
df_clean= df[['Country Name', '2020']]. dropna()
df_clean=df_clean.rename(columns={'2020' : 'Population'})
```

```python
df_top10= df_clean.sort_values(by= 'Population', ascending= False ).head(10)

plt.figure(figsize=(12,6))
ax = sns.barplot(x='Country Name', y='Population', data=df_top10, palette='viridis', edgecolor='black')

# Annotate the bars with values using bar_label
for container in ax.containers:
    ax.bar_label(container, fmt=lambda x: f'{x/1e6:.1f}M', label_type='edge', padding=3)


plt.title('Top 10 Countries by Population (2020)', fontsize=16)
plt.xlabel('Country', fontsize= 14)
plt.ylabel('Population', fontsize= 14)
plt.xticks(rotation= 45, ha='right')
plt.tight_layout()
plt.show()
```

The above represented bar visualization specifies the categorical variables for the population. The graph represent relation between population and countries. Where the dataset doesnt contain gender specification column to perform. So, we go through the Population that specifies the categorical variables which is visualized by using bar chart.

```python
# now histogram distribution for all the countries population (2020)

plt.figure(figsize=(10,6))
sns.histplot(df_clean['Population'], bins=20, color='skyblue', edgecolor= 'black', kde=True)
plt.title('Global Population Distribution (2020)', fontsize=16)
plt.xlabel('Population', fontsize=14)
plt.ylabel('Number of Countries', fontsize=14)
plt.grid(True)
plt.show()
```

The above visualization represents the histogram which means continouos variables for population with Number of Countries for the year 2020.

Accorting to the problem statement we need to visualize the continuous variables for ages for the given dataset. The visualization performed below which specifies the continuous variables for ages and Number of Countries.
